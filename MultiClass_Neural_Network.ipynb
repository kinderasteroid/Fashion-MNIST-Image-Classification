# Build a Multiclass Neural Netowork

```python
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt


from tensorflow.keras.datasets import fashion_mnist

(XTrain,ytrain) , (XTest,ytest) = fashion_mnist.load_data()

plt.imshow(XTrain[0]) #Shoesss
```

#Set the Class names from

```python
lis = ["T-shirt/top" ,"Trouser","Pullover","Dress","Coat","Sandal","Shirt" ,"Sneaker","Bag","Ankle boot"]

plt.imshow(XTrain[3],cmap = plt.cm.binary)
plt.title(lis[ytrain[3]])
```

#Building the Multiclass Classification Model

#Normalize the data

```python
XTrainN = XTrain/255

XTestN = XTest/255
```

```python
tf.random.set_seed(42)

model = tf.keras.Sequential([tf.keras.layers.Flatten(input_shape=(28,28)), tf.keras.layers.Dense(20,activation="relu"),tf.keras.layers.Dense(20,activation="relu"),tf.keras.layers.Dense(10,activation=tf.keras.activations.softmax)])

model.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(),optimizer=tf.keras.optimizers.Adam(),metrics=["accuracy"])

history = model.fit(XTrainN,ytrain,epochs=10 ,validation_data=(XTestN,ytest))
```

```python
pd.DataFrame(history.history).plot()
```
